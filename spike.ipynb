```python
import pandas as pd
import os
```

```python
counts=pd.read_table('ORF_NumReads.tab',sep = " ") #Counts table
```

```python
spike=pd.read_csv('percentmapped_bbmap_volumes.txt', sep='\t')
#spike=pd.read_csv('percentmapped_bbmap.tab', sep=' ') #Spike file with bbmapped reads, total reads, and volume filtered
#spike['VolumeFiltered'] = spike['Volume'] * 0.0625 #Amount of seawater filtered for transcriptomes (1/16th of a filter)
spike
#spike.to_csv('percentmapped_bbmap_volumes.tab', sep=' ', index=True)
```

```python
spikecopies = 500000 #Number of spike copies added to samples (converted from ng)
spikecopies
```

```python
IDs = ['HN0%02d' %i for i in range(1,45)] #Need HN001-HN044
```

```python
#To calculate copies per L for each contig
final = pd.DataFrame()
for i in IDs:
    df = (spike.loc[spike['ID'] == (i)])
    calc = (counts[i] * (spikecopies / (float((df['bbmap-total'])))) * (float(df['Factor']) / (float(df['VolumeFiltered']))))
    colNames = calc.name
    rowNames = list(counts['Name'])
    calc.index = rowNames
    calc.column_name = colNames
    final = final.append(calc)
final_transpose = final.T
print(final_transpose)
final_transpose.to_csv('copiesperL_ORFs_FactorIncluded.csv', sep=' ', index=True)
```
